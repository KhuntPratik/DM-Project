### 23) Bar Chart of Education_Level

```python
edu_counts = bank['Education_Level'].value_counts()
plt.bar(edu_counts.index, edu_counts.values, color='skyblue')
plt.title('Customer Count by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Customers')
plt.show()
```

### 24) Total Transaction Amount over Index

```python
plt.plot(bank['Total_Trans_Amt'].values[:100], color='teal')  # First 100 entries
plt.title('Total Transaction Amount (First 100 Customers)')
plt.xlabel('Index')
plt.ylabel('Total Transaction Amount')
plt.grid(True)
plt.show()
```

### 25)  Histogram of customer ages

```python

plt.figure(figsize=(8, 5))
plt.hist(bank['Customer_Age'], bins=10, color='cornflowerblue', edgecolor='black')
plt.title('Distribution of Customer Age')
plt.xlabel('Age')
plt.ylabel('Number of Customers')
plt.grid(True)
plt.show()
```
